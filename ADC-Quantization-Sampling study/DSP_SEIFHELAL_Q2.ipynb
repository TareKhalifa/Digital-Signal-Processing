```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Question 2 Part A : a 5hz sin wave running for 1 second

#for orginal sin wave
time = 1.00 
frame_rate= 20000 # increasing the range of frame rate to be 20000 per second to make it look like analoug signal
frequency = 5 # the frequency of the signal
sin_range = np.arange(0,time, (1.0/frame_rate)) # the points on the x axis for plotting
sin_funct = [ np.sin(2*np.pi*frequency * (i/time)) for i in sin_range] # orginal sin wave 


#Question 2 Part B: quantized sin wave
#quantization


def analog_signal (time_point):
    return np.sin(2*np.pi*frequency * (time_point/time))

quantizing_bits     = 8  # bit rate for quantization
quantizing_levels   = (2 ** quantizing_bits - 1 ) / 2
quantizing_step     = 1. / (quantizing_levels)
mod_frame_rate= 40.0 # frame rate used with sampling
mod_sin_range = np.arange(0,time, (1.0/mod_frame_rate)) # the points on the x axis for plotting
mod_sin_funct = analog_signal(mod_sin_range) # digital sin wave sampled with mod_frame_rate
quantized_mod_sin_funct= np.round (mod_sin_funct / quantizing_step) * quantizing_step #quantized modified sin wave




#Question 2 Part C: qunatization error avg and standard deviation
errors= (mod_sin_funct - quantized_mod_sin_funct )*quantizing_levels
error_mean=np.mean(errors)
error_std= np.std(errors)
print ("The Quantization Error Avg = %f" %error_mean)
print ("The Error standard deviation = %f  LSB" %error_std)

#Question 2 Part D: Theoritical qunatization error avg and standard deviation
print ("The Theortical Quantization Error Avg = 0")
theoritical_std=(1.0/np.sqrt(12))
print ("The Theoritical standard deviation = %f  LSB" %theoritical_std)
print ("Difference between theortical Standard Deviation and experimental standard Deviatio = %f  LSB"% (theoritical_std-error_std))

plt.title("Sin Wave 5 Hz ")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.plot(sin_range,sin_funct, '-')
plt.stem (mod_sin_range, quantized_mod_sin_funct, linefmt='r-', markerfmt='ro', basefmt='r-');
plt.plot(mod_sin_range,mod_sin_funct, 'ro')
plt.show()
```

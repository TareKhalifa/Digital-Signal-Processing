```python
import matplotlib.pyplot as plt
import numpy as np
import wave 
```

```python

change_rate = 0.25 # plug in the fatctor you want to stretch
#retrive the orginal rate and signal of the file. 
read_file = wave.open('test_q1.wav', 'rb') #read the file
original_rate=read_file.getframerate() # get number of audio frames 
original_signal = read_file.readframes(-1) # read all frames of audio as a string of bytes

write_file = wave.open('updated_test_q1.wav', 'wb') #the new file 
write_file.setnchannels(1) # set the channels to be mono =1 
write_file.setsampwidth(2) # set the sample width to be 2 bytes
write_file.setframerate(original_rate*(1/change_rate)) # set the frame rate to be a multiple of the orginal rate
write_file.writeframes(original_signal) # write the audio frames 
```

```python
#original wav file
read_file = wave.open('test_q1.wav','r')
signal = read_file.readframes(-1)
signal = np.fromstring(signal, 'Int16')
fs = read_file.getframerate()
Time=np.linspace(0, len(signal)/fs, num=len(signal))
plt.figure(1)
plt.title('Orginal Wave file')
plt.plot(Time,signal)
plt.show()


# updated file 
write_file = wave.open('updated_test_q1.wav','r')
signal = write_file.readframes(-1)
signal = np.fromstring(signal, 'Int16')
fs = write_file.getframerate()
Time=np.linspace(0, len(signal)/fs, num=len(signal))
plt.figure(1)
plt.title('Updated Wave file')
plt.plot(Time,signal)
plt.show()
```

```python
#exponential_signal
def exp_signal(t):
    x=0.0
    if(t>=0):
        x= np.exp(-t)
    return x
vect_exp_signal= np.vectorize(exp_signal)


x_points = np.linspace(-2.5, 2.5, 100) #100 per 5 seconds
samples=vect_exp_signal(x_points)
plt.plot(x_points, samples, 'bo')
plt.plot(x_points, samples, 'r-')
plt.show()
```

```python
#Question2  part 1

print ("Generated signal represented as blue dots: ")
plt.plot( samples, 'bo')
plt.plot(samples, 'r-')

plt.show()
dft_points2=DFT(samples)
print ("100 point DFT ")
dft_points_magnitude2= np.abs(dft_points2)
plt.plot(dft_points_magnitude2, 'r')
plt.show()


idft_points2=IDFT(np.array(dft_points2))
print ("100 point IDFT  ")
idft_points_magnitude2= np.array(idft_points2).real
plt.plot(idft_points_magnitude2)
plt.show()
```

```python
#question2 part 2 
x_points = np.linspace(-2.5, 2.5, 100) #100 per 5 seconds
samples=vect_exp_signal(x_points)
samples=np.append(samples,np.zeros(100))
print ("Generated signal represented as blue dots: ")
plt.plot( samples, 'bo')
plt.plot( samples, 'r-')

plt.show()

dft_points2=DFT(samples)
print ("200 point DFT ")
dft_points_magnitude2= np.abs(dft_points2)
plt.plot(dft_points_magnitude2, 'r')
plt.show()


idft_points2=IDFT(np.array(dft_points2))
print ("200 point IDFT  ")
idft_points_magnitude2= np.array(idft_points2).real
plt.plot(idft_points_magnitude2)
plt.show()
```

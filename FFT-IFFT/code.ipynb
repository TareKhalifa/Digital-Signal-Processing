```python
import numpy as np
import scipy.fftpack
import scipy.io.wavfile as wf
import matplotlib.pyplot as plt
from utilities import *
```

```python
#Question 1
    # reading the input file that contains my name
_,name_file=wf.read("seif.wav")
data_name_file,end=np.array(sample(name_file,32000))
plt.title("The name file data graph \n")
plt.plot(data_name_file, 'r-')
plt.show()
    #applying the fft
fft_name_file=FFT(data_name_file).fft()
mag_fft=np.abs(fft_name_file)
plt.title("All points FFT and magnitude part graph \n")
plt.plot(mag_fft, 'r-')
plt.show()
    #applying the ifft
ifft_name_file= IFFT(fft_name_file).ifft()[:end*2]
plt.title("Applying IFFT graph \n")
plt.plot(ifft_name_file, 'r-')
plt.show()
    #comapring the ifft output to the original file by drwaing their graphs on top of each others 
plt.title("Comapring ifft to original \n")
plt.plot(data_name_file, 'r-')
plt.plot(ifft_name_file)
plt.show()

    #writing back to the file
wf.write("seif_new.wav", ifft_name_file.shape[0], ifft_name_file)
```

```python

#Question 2
 # reading the input file that contains my name and whistle
_,name_file=wf.read("seif+whistle3.wav")
end=name_file.shape[0]
while (name_file.shape[0]<65000)  :
    name_file=np.append(name_file,0)    
data_name_file=np.array(name_file)
plt.title("The name+whistle file data graph \n")
plt.plot(data_name_file, 'r-')
plt.show()
   #applying the fft
fft_name_file=FFT(data_name_file).fft()
mag_fft=np.abs(fft_name_file)
plt.title("All points FFT and magnitude part graph \n")
plt.plot(mag_fft, 'r-')
plt.show()
    #applying low pass filter
filtered_input=lowpass_filter(fft_name_file)
mag_filtered= np.abs(filtered_input)
plt.title("after low pass filter graph \n")
plt.plot(mag_filtered, 'r-')
plt.show()

    #applying the ifft
ifft_name_file= IFFT(filtered_input).ifft()[:65536*2]
plt.title("Applying IFFT graph \n")
plt.plot(ifft_name_file, 'r-')
plt.show()
    #comapring the ifft output to the original file by drwaing their graphs on top of each others 
plt.title("Comapring ifft to original \n")
plt.plot(data_name_file, 'r-')
plt.plot(ifft_name_file)
plt.show()

    #writing back to the file
wf.write("seif+whistle_new.wav", ifft_name_file.shape[0], ifft_name_file)
```
